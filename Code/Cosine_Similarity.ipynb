```python
import os
import pandas as pd
import numpy as np
import re
import nltk
import string
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_colwidth', 0)
```

```python
# data = [[filename for filename in os.listdir(os.path.join(path, directory)) if filename[-4:] == '.txt'] for directory in os.listdir(path) if os.path.isdir(os.path.join(path, directory))]
topics = {}
path = 'C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/YT_Corpus [All]/YT_Corpus [2016-2017]'
for directory in os.listdir(path):
    if os.path.isdir(os.path.join(path, directory)):
        topics[directory]={}
        for filename in os.listdir(os.path.join(path, directory)):
            if filename == 'LDAGibbs 5 TopicsToTerms.csv':
                topics_df = pd.read_csv(os.path.join(path, directory, filename), usecols=[1,2,3,4,5])
                for col in topics_df.columns:
                    topics[directory][directory + " - " + col] = ' '.join(topics_df[col])
```

```python
reports = {}
path = 'C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Text_Corpus'
for directory in os.listdir(path):
    if os.path.isdir(os.path.join(path, directory)):
        reports[directory]={}
        for filename in os.listdir(os.path.join(path, directory)):
            if filename[-4:] == '.txt':
                with open(os.path.join(path, directory, filename)) as f:
                    file = f.read()
                    reports[directory][filename] = file
```

```python
reports['ACCENTURE_PLC_1467373']
```

```python
ps = nltk.PorterStemmer()
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')

reports_corpus = pd.DataFrame.from_dict(reports['ACCENTURE_PLC_1467373'], orient='index')
reports_corpus = reports_corpus[0].apply(lambda x: re.sub("[^\x30-\x7A]+", " ", x))
reports_corpus = reports_corpus.apply(lambda x: " ".join([ps.stem(word) for word in x.split(' ') if word not in string.punctuation and word not in stopwords]))
reports_corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(max_features=100)
tfidf_vect_fit = tfidf_vect.fit(reports_corpus)

tfidf_reports_corpus = tfidf_vect_fit.transform(reports_corpus)

from sklearn.decomposition import TruncatedSVD

tfidf_reports_corpus = pd.DataFrame.sparse.from_spmatrix(tfidf_reports_corpus)
tfidf_reports_corpus
```

```python
topics_corpus = pd.DataFrame.from_dict(topics)
topics_corpus = topics_corpus[topics_corpus.columns].apply(lambda x: ','.join(x.dropna().astype(str)), axis=1)

topics_corpus
```

```python
tfidf_vect = TfidfVectorizer(max_features=2000)
tfidf_vect_fit = tfidf_vect.fit(topics_corpus)

tfidf_topics_corpus = tfidf_vect_fit.transform(topics_corpus)

svd = TruncatedSVD(n_components=100, n_iter=10).fit(tfidf_topics_corpus)
lsa_topics_corpus = svd.transform(tfidf_topics_corpus)

lsa_topics_corpus = pd.DataFrame(lsa_topics_corpus)
lsa_topics_corpus
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

cosine_results = pd.DataFrame(index=reports_corpus.index, columns=topics_corpus.index)

for r in range(len(tfidf_reports_corpus)):
    for t in range(len(lsa_topics_corpus)):
        cosine_results.iloc[r, t] = cosine(tfidf_reports_corpus.iloc[r], lsa_topics_corpus.iloc[t])

cosine_results = cosine_results.apply(pd.to_numeric)
cosine_results
```
