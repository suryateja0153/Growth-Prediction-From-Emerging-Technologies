```python
import os
import shutil
import pandas as pd
from os import listdir
from os.path import isfile, join

onlyfiles = [f for f in listdir('C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Temp_Files_2') if isfile(join('C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Temp_Files_2', f))]
newOp = list({x.replace(".xlsx", "") for x in onlyfiles})
temp = newOp
temp
```

```python
for row in temp:
    directory = str(row)
    parent_dir = 'C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Text_Corpus/'
    path = os.path.join(parent_dir, directory)
    os.mkdir(path)
    
    base_path = 'C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Temp_Files_2/' + row + '.xlsx'   
    shutil.copy2(base_path,path + '/')
```

```python
for rows in temp:
    df = pd.read_excel('C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Temp_Files_2/' + rows + '.xlsx', sheet_name='Sheet1', index=0)

    del df['year']
    del df['final_url']

    if (df.columns[0] == 'acc-no'):
        df.rename(columns={'acc-no': 'acc_no'}, inplace=True)

    selected_columns1 = df[["acc_no","text_business"]]
    df1 = selected_columns1.copy()

    selected_columns2 = df[["acc_no","text_risk"]]
    df2 = selected_columns2.copy()

    file = 'C:/Users/surya/Downloads/Career/M.S. [2020-22]/3.Spring_2021/[ISM6905.001S21.12437 Independent Study]/Text_Corpus/' + rows + '/{}.txt'
    for i, ref in df1.iterrows():
        try:
            with open(file.format(ref['acc_no']+'_Business'), 'w') as f:
                f.write(str(ref['text_business']))
        except:
            break

    for i1, ref1 in df2.iterrows():
        try:
            with open(file.format(ref1['acc_no']+'_Risk'), 'w') as f1:
                f1.write(str(ref1['text_risk']))
        except:
            break
```
